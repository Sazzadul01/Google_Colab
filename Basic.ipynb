```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/churn-bigml-80.csv")
```

```
df.shape
```

```
# missing value Check
df.isnull().sum()
# if any  (df = df.dropna())
```

```
# Duplicate_rows check
duplicate_rows = df[df.duplicated()]
print(f"\n No of duplicate rows: {duplicate_rows.shape[0]}")
# if any duplicate (df = df.drop_duplicates())
```

```
#standarize inconsistent data froms
print("Before Standardization")
print(df.columns)
df.columns = df.columns.str.lower().str.replace(" ","_")
print("\nAfter Standardization")
print(df.columns)
```

```
df.describe()
```

```
# Histogram
import matplotlib.pyplot as plt
df['total_day_minutes'].hist(bins=20)
plt.title("Total Day Minutes")
plt.xlabel("Total Day Minutes")
plt.ylabel("Frequency")
plt.show()
```

```
# Boxplot
df.boxplot(column='total_day_minutes')
plt.title("Total Day Minutes")
plt.show()
```

```
# Scatter plot
plt.scatter(df['total_day_minutes'],df["total_day_charge"])
plt.title("Day Minutes vs Day Charge")
plt.xlabel("Total Day Minutes")
plt.ylabel("Total Day Charge")
plt.show()
```

```
# Correlation vs Numerical
import seaborn as sn
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sn.heatmap(correlation, cmap='coolwarm')
plt.title("Correlation")
plt.show()
```

```
# Bar
df['churn'].value_counts().plot(kind='bar')
plt.title("churn Distribution")
plt.xlabel("churn")
plt.ylabel("count")
plt.legend(["Cusomers"])
plt.show()
```

```
# Line
plt.plot(df['total_day_minutes'])
plt.title("Total day minutes trend")
plt.xlabel("index")
plt.ylabel("Total day minutes")
plt.legend(["Day Minutes"])
plt.show()
```

```
# Scatter
plt.scatter(df['total_day_minutes'],df["total_day_charge"])
plt.title("Day Minutes vs Day Charge")
plt.xlabel("Total Day Minutes")
plt.ylabel("Total Day Charge")
plt.legend(["Customers"])
plt.show()
```
